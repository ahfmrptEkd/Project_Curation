# í›„ì²˜ë¦¬ ì‹œì‘

```python
# í•¨ìˆ˜ íŒŒíŠ¸

# ë³„ì ì„ ìˆ«ìë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def convert_rating_to_numeric(rating):
    """
    ë³„ì  ë¬¸ìì—´ì„ ìˆ«ìë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
    
    Args:
        rating (str): ë³„ì  ë¬¸ìì—´ (ì˜ˆ: "â­â­â­â­", "â­â­â­")
        
    Returns:
        int: ë³€í™˜ëœ ìˆ«ì (1-5)
    """
    rating_mapping = {
        'â­': 1,
        'â­â­': 2,
        'â­â­â­': 3,
        'â­â­â­â­': 4,
        'â­â­â­â­â­': 5
    }
    return rating_mapping.get(rating, 0)  # ë§¤í•‘ë˜ì§€ ì•Šì€ ê²½ìš° 0 ë°˜í™˜
```

```python
print(data2025_cleaned['Rating'].value_counts(), "\n")
print(data2024_cleaned['Rating'].value_counts(), "\n")
print(data2023_cleaned['Rating'].value_counts(), "\n")



# Rating ì»¬ëŸ¼ì„ ìˆ«ìë¡œ ë³€í™˜
data2024_cleaned['Rating_Numeric'] = data2024_cleaned['Rating'].apply(convert_rating_to_numeric)
data2023_cleaned['Rating_Numeric'] = data2023_cleaned['Rating'].apply(convert_rating_to_numeric)
data2025_cleaned['Rating_Numeric'] = data2025_cleaned['Rating'].apply(convert_rating_to_numeric)

# ë³€í™˜ ê²°ê³¼ í™•ì¸
print("ë³€í™˜ëœ ë³„ì  ë¶„í¬:")
print("2025ë…„:", data2025_cleaned['Rating_Numeric'].value_counts().sort_index())
print("2024ë…„:", data2024_cleaned['Rating_Numeric'].value_counts().sort_index())
print("2023ë…„:", data2023_cleaned['Rating_Numeric'].value_counts().sort_index())
```
