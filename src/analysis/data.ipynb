# 후처리 시작

```python
# 함수 파트

# 별점을 숫자로 변환하는 함수 정의
def convert_rating_to_numeric(rating):
    """
    별점 문자열을 숫자로 변환하는 함수
    
    Args:
        rating (str): 별점 문자열 (예: "⭐⭐⭐⭐", "⭐⭐⭐")
        
    Returns:
        int: 변환된 숫자 (1-5)
    """
    rating_mapping = {
        '⭐': 1,
        '⭐⭐': 2,
        '⭐⭐⭐': 3,
        '⭐⭐⭐⭐': 4,
        '⭐⭐⭐⭐⭐': 5
    }
    return rating_mapping.get(rating, 0)  # 매핑되지 않은 경우 0 반환
```

```python
print(data2025_cleaned['Rating'].value_counts(), "\n")
print(data2024_cleaned['Rating'].value_counts(), "\n")
print(data2023_cleaned['Rating'].value_counts(), "\n")



# Rating 컬럼을 숫자로 변환
data2024_cleaned['Rating_Numeric'] = data2024_cleaned['Rating'].apply(convert_rating_to_numeric)
data2023_cleaned['Rating_Numeric'] = data2023_cleaned['Rating'].apply(convert_rating_to_numeric)
data2025_cleaned['Rating_Numeric'] = data2025_cleaned['Rating'].apply(convert_rating_to_numeric)

# 변환 결과 확인
print("변환된 별점 분포:")
print("2025년:", data2025_cleaned['Rating_Numeric'].value_counts().sort_index())
print("2024년:", data2024_cleaned['Rating_Numeric'].value_counts().sort_index())
print("2023년:", data2023_cleaned['Rating_Numeric'].value_counts().sort_index())
```

```python
data2023_cleaned['Trope']

# 특정 키워드가 포함된 행들을 필터링하여 확인
keywords = ['onceler', 'omegaverse', 'gangbang', 'Dom/Sub', 'cars']

# 각 연도별 데이터에서 키워드가 포함된 행들을 찾기
for year, data in [('2025', data2025_cleaned), ('2024', data2024_cleaned), ('2023', data2023_cleaned)]:
    print(f"\n=== {year}년 데이터에서 키워드가 포함된 행들 ===")
    
    # 키워드가 포함된 행들을 찾기
    filtered_rows = data[data['Trope'].str.contains('|'.join(keywords), case=False, na=False)]
    
    if len(filtered_rows) > 0:
        print(f"총 {len(filtered_rows)}개 행 발견:")
        for idx, row in filtered_rows.iterrows():
            print(f"  - {row['Book Title']} | {row['Author']} | {row['Trope']} | {row['Rating_Numeric']}점")
    else:
        print("해당 키워드가 포함된 행이 없습니다.")
```

```python
# 키워드가 포함된 행들을 찾아서 삭제
rows_to_remove = data2023_cleaned[data2023_cleaned['Trope'].str.contains('|'.join(keywords), case=False, na=False)]

print(f"삭제할 행 수: {len(rows_to_remove)}")
print("삭제할 행들:")
for idx, row in rows_to_remove.iterrows():
    print(f"  - {row['Book Title']} | {row['Author']} | {row['Trope']}")

# 해당 행들을 삭제
data2023_cleaned = data2023_cleaned[~data2023_cleaned['Trope'].str.contains('|'.join(keywords), case=False, na=False)]

print(f"\n삭제 후 데이터 크기: {data2023_cleaned.shape}")
print(f"삭제 전 크기와 비교: {len(data2023_cleaned) + len(rows_to_remove)} -> {len(data2023_cleaned)}")
```
